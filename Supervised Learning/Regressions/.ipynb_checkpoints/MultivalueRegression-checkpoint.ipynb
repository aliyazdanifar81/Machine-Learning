```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import minmax_scale
from sklearn.metrics import mean_squared_error
from sklearn.decomposition import PCA
from load_data import LoadModel
import matplotlib.pyplot as plt
import numpy as np
```

```python
x, y = LoadModel().load_house(feature=4)
```

```python
x, y = minmax_scale(x), minmax_scale(y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, shuffle=True, random_state=42)
```

```python
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
model.coef_, model.intercept_
```

```python
pca = PCA(n_components=1)
pca = pca.fit_transform(x)
```

```python
pca3d = PCA(n_components=2)
pca3d = pca3d.fit_transform(x)
```

```python
ax = plt.axes(projection ='3d')
z = y
x_ = pca3d[:, 0]
y_ = pca3d[:,1]
ax.scatter(x_, y_, z, 'green')
```

```python
plt.scatter(pca, y)
```

```python
pre = model.predict(x_test)
```

```python
mean_squared_error(pre, y_test)
```

```python
train_accuracy, test_accuracy = model.score(x_test, y_test), model.score(x_train, y_train)
print(f"train_accuracy: {train_accuracy}\ntest_Accuracy: {test_accuracy}")
```
